```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import sklearn
from sklearn import preprocessing
import math
```

```python
df = pd.read_csv("..//resources//csv//data//biometrics_juan.csv")
```

```python
colnames = ['HR', 'HRV', 'MicroSiemens']
df_std = df[colnames]
names = df_std.columns

scaler = preprocessing.StandardScaler()
scaled_df = scaler.fit_transform(df_std)

scaled_df = np.append(scaled_df, df[['ArousalMean']], axis=1)

scaled_df = pd.DataFrame(scaled_df, columns=['HR', 'RR', 'HRV', 'MicroSiemens', 'SCR', 'ArousalMean'])
scaled_df.head()

# sns.pairplot(scaled_df, height=3.5);
# plt.show()
```

```python
# remover las mediciones sin estímulo
filterArousal = scaled_df['ArousalMean'] > 0
filtered1 = scaled_df[filterArousal]

# remover las mediciones para las que el sam no coincidió
# filterSam = filtered1['MatchesSam'] == True
# filtered = filtered1[filterSam]

filtered1.head()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(filtered1.corr(), annot=True, fmt=".2f", cmap='YlGnBu')
plt.show()
```

```python
X = filtered1[['MicroSiemens', 'HR', 'HRV' ]].values
y = filtered1['ArousalMean'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False, random_state=42)

y_train, y_test
```

```python
# Para poder usar un clasificador, los datos de la variable target deben ser discretos.
#Los convertimos a True-False (Aroused-NotAroused) con el fin de poder entrear un clasificador binario
y_train_aroused = (y_train > 5)
```

```python
y_train_aroused
```

```python
from sklearn.linear_model import SGDClassifier

clf = SGDClassifier(random_state = 0)
clf.fit(X_train, y_train_aroused)
```

# Prediction

```python
clf.predict(X[145].reshape(1, -1))
```

# Performance Measuring

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(clf, X_train, y_train_aroused, cv=3, scoring='accuracy')
```

# Confusion Matrix

```python
from sklearn.model_selection import cross_val_predict
```

```python
y_train_pred = cross_val_predict(clf, X_train, y_train_aroused, cv=3)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_train_aroused, y_train_pred)
```

```python
pd.DataFrame(confusion_matrix(y_train_aroused, y_train_pred),
             columns=pd.MultiIndex.from_product([['Prediction'], ["Negative", "Positive"]]),
             index=pd.MultiIndex.from_product([["Actual"], ["Negative", "Positive"]]))
```

```python
from sklearn.metrics import precision_score, recall_score
```

```python
precision_score(y_train_aroused, y_train_pred)
```
