```python
import pickle
import pandas as pd
from sklearn import preprocessing
import matplotlib.pyplot as plt
import numpy as np
```

```python
model = pickle.load(open('./models/knn-hr-skin-model', 'rb'))
# result_val = model.score(xval, yval)
# result_test = model.score(xtest, ytest)
```

```python
df = pd.read_csv("..//resources//csv//data//biometrics-juan.csv")
df.head()
```

```python
def map_to_0_and_1(arousal):
    return [0] if arousal < 5 else [1]
```

```python
df_aux = df.copy()
columns_to_standarize = ['HR', 'HRV', 'MicroSiemens']

# remover las mediciones sin estÃ­mulo
filterArousal = df_aux['ArousalMean'] > 0
df_aux = df_aux[filterArousal]

is_aroused = np.array(list(map(map_to_0_and_1, df_aux['ArousalMean'])))

scaler = preprocessing.StandardScaler()

standarized_columns = scaler.fit_transform(df_aux[columns_to_standarize])

df_to_predict = pd.DataFrame(standarized_columns, columns=['HR', 'HRV', 'MicroSiemens'])

df_to_predict['ArousalMean'] = is_aroused
df_to_predict['PhaseName'] = np.array(df_aux['PhaseName'])
df_to_predict['MatchesSam'] = np.array(df_aux['MatchesSam'])
df_to_predict['TimeStamp'] = np.array(df_aux['TimeStamp'])

df_to_predict
```

```python
y_pred = model.predict(df_to_predict[['HR', 'HRV', 'MicroSiemens']])
```

```python
plt.plot(y_pred)
plt.show()
```

```python
model.score(df_to_predict[['HR', 'HRV', 'MicroSiemens']], df_to_predict[['ArousalMean']])
```

```python
df_to_predict['ArousalPred'] = y_pred

df_to_predict.to_csv(r'./prediction.csv', index = False)
```
